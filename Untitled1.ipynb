```python
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.grid(True)
print(plt.axis())
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, -x**2)
ax.plot(x, -x**3)
ax.plot(x, -2*x)
ax.plot(x, -2**x)
ax.set_xlabel('x = np.arange(3)')
ax.set_ylabel('y = f(x)')
ax.set_title('Simple Plot Demo')
ax.grid(True)
plt.show()
```
